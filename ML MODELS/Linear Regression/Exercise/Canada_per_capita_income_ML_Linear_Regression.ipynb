```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
capita_df = pd.read_csv('canada_per_capita_income.csv')
capita_df.head()
```

```python
capita_df = capita_df.rename(columns = {'per capita income (US$)':'per_capita'})
capita_df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita')
plt.scatter(capita_df.year,capita_df.per_capita,color='red',marker='+')
plt.show()
```

```python
year_df = capita_df.drop('per_capita',axis='columns')
year_df.head()
```

```python
per_capita = capita_df.per_capita
```

```python
reg = linear_model.LinearRegression()
reg.fit(year_df.values,per_capita)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2022]])
```
