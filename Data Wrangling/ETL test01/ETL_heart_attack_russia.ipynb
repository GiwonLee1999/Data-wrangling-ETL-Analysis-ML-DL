```python
data = pd.read_csv('heart_attack_russia_youth_vs_adult.csv')
len(data)
```

```python
data.isna() # false means no missing data
```

```python
missing_data = data.isna().sum()
missing_data
```

```python
# Based off of this, we can see that there are no missing values in the dataset except for the Alcohol consumption
data['Alcohol_Consumption']
data['Alcohol_Consumption'].fillna('None', inplace=True)
data['Alcohol_Consumption'].isna().sum()
# Now we have no missing values in the Alcohol Consumption column
```

```python
# Sample data is created
sample_data = data.dropna() # drops the rows with missing values
len(sample_data)
```

```python
sample_data.head()
```

```python
missing_count = sample_data.isna().sum()
missing_value_percentage = missing_count / len(sample_data) * 100
# missing_value_percentage
region_group = sample_data.groupby(['Region']).size()
gender_group = sample_data.groupby(['Gender']).size()
age_group = sample_data.groupby(['Age']).size()
# Here we can see the number of people in each age group and 28 year olds are the most heart attacks.
# age_group.sort_values(ascending=False)
```
